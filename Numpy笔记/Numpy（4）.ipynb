```python
import numpy as np
#（1）访问数组切片
arr1 = np.arange(10)
print(arr1)
```

```python
#从 x 切到 y的几种解法(中间用冒号)
#注意！切片这里的索引不太一样。对应位置：  （0）a （1）b （2）c。简单理解第几个元素就代表它和它之前
print(arr1[1:4])
#从x切到y
print(arr1[1:])
#第二个空着代表切到末尾
print(arr1[:5])
#第一个空着代表从头切到y
```

```python
#带负号的情况
print(arr1[2:-2])
#切除开头和结尾两个
```

```python
#隔几个元素切
print(arr1[ : :2])
#每隔两个元素采样
print(arr1[1:-2:2])
#注意！有负数率代表切除，切掉，不要了！本例先切除对应元素。再每隔两个元素取样
```

```python
#（2）接下来是访问矩阵切片
arr2 = np.arange(0,20).reshape(4,5)
print(arr2)
```

```python
print(arr2[1:3,1:-1])
#先取一到三行，再把第一列和最后一列切除不要！
```

```python
#跳跃采样
print(arr2[::3,::2])
#每隔几行几列取。这里的索引注意从o开始，对应的是第一行元素之前。详看之前
```

```python
#(3)提取矩阵的行
print(arr2)
print("result：")
print(arr2[2,:])
#提取第二行
```

```python
#提取一到二行，因为区间是左闭右开
#最好不要简写成arr[2]
print(arr2[1:3,])
#观察结果的括号数量，会发现输出的是一个矩阵。
```

```python
#(4)提取列向量
print(arr2[:,2])
#提取第二列，观察结果可发现结果是向量！！！
```

```python
print(arr2[:,1:3])
#提取一到2列，左闭右开
```

```python
#(4)数组切片仅是视图。意思书Numpy的切片不会创建新变量
#但也可以使用copy创建拷贝切片
copy = arr1[:4].copy()
print(copy)
print(arr1)
```
