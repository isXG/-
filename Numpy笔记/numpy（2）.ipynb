```python
import numpy as np
```

```python
#创建一维数组--向量
arr1 = np.array([1,2,3])
print(arr1)
```

```python
#创建二维数组--行矩阵
arr2 = np.array([[1,2,3]])
print(arr2)
```

```python
#创建二维数组--列矩阵
arr3 = np.array([[1],[2],[3]])
print(arr3)
```

```python
#使用arange创建递增数组
#只传入一个参数，打印从0到n
arr4 = np.arange(10)
print(arr4)
```

```python
#传入两个参数打印n1到n2
arr5 = np.arange(10,20)
print(arr5)
```

```python
#传入三个参数，从n1开始，到n2结束，步长为n3
arr6 = np.arange(1,33,4)
print(arr6)
```

```python
#使用np.zeros()函数以及np.ones()函数创建同值数组
#对于，zeros（）只传入一个数字，返回一个向量。
arr_1 = np.zeros(3)
print (arr_1)
#传入行矩阵,不同之处要用括号括起来：
arr_2 = np.ones((1,3))
print(arr_2)
#全一数组同理
#具体的可以用np.ones创建形状在在前面乘以数字
arr_3 = 3.14 * np.ones((1,3))
print(arr_3)
```

```python
#使用np.random创建随机数组
#该系列函数很多，主要讲解四个
#第一个random：创建0-1均匀分布的浮点型随机数组。
#下式第一个random代表np中的模块random，第二个random代表模块random中的函数，5代表形状为5的向量。
arr_3 = np.random.random(5)
print(arr_3)
```

```python
#也可以生产几行几列的矩阵例如：
arr_4 = np.random.random((2,9))
print(arr_4)
```

```python
#该函数并非只能创建0-1.例如：
#如果想创建60-100范围内均匀分布的3行3列随机数组，可输入：
arr_5 = 40*np.random.random((3,5))+60;
print(arr_5)
```

```python
#第二个randint：创建整数型随机数组
#随机生产一个
arr_6 = np.random.randint(10,100)
print(arr_6)
```

```python
#也可以好几个，也可以矩阵,形状有第三个参数决定
arr_7 = np.random.randint(10,100,4)
arr_8 = np.random.randint(10,100,(3,6))
print(arr_7)
print(arr_8)
```

```python
#第三个：使用normal创建服从正态分布的随机数组,第三个依旧形状，不赘述
arr_9 = np.random.normal(0,1,9)
#本例中均值为0，标准差为1，可直接使用np.random.randn实现
#normal可以指定均值和标准差，默认为0
print(arr_9)
```
